### Q2

```python
import inspect
import logging
logging.basicConfig(filename='Week-05(Exception Handling Assignment-2).log',level=logging.INFO)
logging.info('The Q2 is being executed.')
print("The class hierarchy for built-in exceptions is:")
inspect.getclasstree(inspect.getmro(Exception))
def classtree(cls, indent=0):
    print('.' * indent, cls.__name__)
    for subcls in cls.__subclasses__():
        classtree(subcls, indent + 3)
logging.info('The Q2 is executed.')
classtree(Exception)
```

### Q3

```python
# The ArithmeticError exception class serves as a base class for those built-in exceptions that are raised
# due to various arithmetic errors like OverflowError, ZeroDivisionError, FloatingPointError.
```

```python
# Example(1) : ZeroDivisionError is caused when a number is divided by zero.
try:
    logging.info('The Q3(ZeroDivisionError) is being executed.')
    a=int(input('Enter numerator : '))
    b=int(input('Enter denominator : '))
    c=a/b
    print(f'Output : {c}')
    logging.info('The Q3(ZeroDivisionError) is executed.')
except ZeroDivisionError as e:
    print(e)
    logging.error(f'There is a error : {e}')
```

```python
# Example(2) : OverflowError is a type of exception that occurs when an operation produces a result that
# exceeds the maximum value that can be stored in memory. 

try:
    import math
    logging.info('The Q3(OverflowError) is being executed. ')
    print(math.exp(1000))
    logging.info('The Q3(OverflowError) executed. ')
except OverflowError as e:
    print(e)
    logging.error(f'There is a error : {e}')
```

### Q4

```python
# LookupError Exception is the Base class for errors raised when something can’t be found. The base class
# for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: 
# IndexError, KeyError.
```

```python
# IndexError
try:
    logging.info('The Q4(IndexError) is bieng executed.')
    a=[10,20,30,40,50]
    print(a[5])
    logging.info('The Q4(IndexError) executed.')
except IndexError as e:
    print(e)
    logging.error(f'There is a error : {e}')
```

```python
# KeyError
try:
    logging.info('The Q4(KeyError) is bieng executed.')
    a={'name':'Pratik','email-id':'pratik@gmail.com'}
    print(a['mobile_no'])
    logging.info('The Q4(KeyError) executed.')
except KeyError as e:
    print(e)
    logging.error(f'There is not any key named : {e}')
```
